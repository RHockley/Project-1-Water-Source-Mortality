```python
gdppc = "tiny_df3.csv"

gpc_read = pd.read_csv(gdppc, encoding = "utf-8")
finalGDP = pd.DataFrame(gpc_read)

finalGDP.head()
```

```python
gnd_df
```

```python
gnd_df = gnd_df.dropna(how="any")
```

```python
gnd_df.dtypes
```

```python
gnd_df.astype({"Year_y": 'int64'})
```

```python
clean_df = gnd_df.loc[((gnd_df["Year_x"] == 2015) & (gnd_df["Year_y"] == 2015)) |((gnd_df["Year_x"] == 2016) & (gnd_df["Year_y"] == 2016)) | ((gnd_df["Year_x"] == 2017) & (gnd_df["Year_y"] == 2017)) | ((gnd_df["Year_x"] == 2018) & (gnd_df["Year_y"] == 2018))| ((gnd_df["Year_x"] == 2019) & (gnd_df["Year_y"] == 2019)) ]
```
