```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
dirrahea_deaths = "../sources/dirrahea_deaths.csv"
```

```python
# Read our Data file with the pandas library
dirrahea_deaths_df = pd.read_csv(dirrahea_deaths)
```

```python
# Show just the header
dirrahea_deaths_df.head(50)
```

```python
dirrahea_deaths_reduced = dirrahea_deaths_df[["Entity", "Code", "Year", "Deaths - Cause: Diarrheal diseases - Risk: Unsafe water source - Sex: Both - Age: Under 5 (Number)", "Deaths - Cause: Diarrheal diseases - Risk: No access to handwashing facility - Sex: Both - Age: Under 5 (Number)"]]
dirrahea_deaths_reduced.head(30)
```

```python
dirrahea_deaths_reduced_loc = dirrahea_deaths_reduced.loc[dirrahea_deaths_reduced['Year'].isin(["2015", "2016", "2017", "2018", "2019"])]
dirrahea_deaths_reduced_loc
```

```python
dirrahea_deaths_reduced_loc["Entity"].unique()
```

```python
dirrahea_deaths_most_developed = dirrahea_deaths_reduced_loc.loc[dirrahea_deaths_reduced_loc['Entity'].isin(["Monaco", "Luxembourg", "Bermuda", "Switzerland", "Norway"])]
dirrahea_deaths_most_developed
```

```python
dirrahea_deaths_most_developed_avg = dirrahea_deaths_most_developed.groupby(['Entity', 'Code']).mean()
dirrahea_deaths_most_developed_avg

#The average of the five most developed countries dataframe.
```

```python
dirrahea_deaths_least_developed = dirrahea_deaths_reduced_loc.loc[dirrahea_deaths_reduced_loc['Entity'].isin(["Madagascar", "Mozambique", "Malawi", "Somalia", "Burundi"])]
dirrahea_deaths_least_developed
```

```python
dirrahea_deaths_least_developed_avg = dirrahea_deaths_least_developed.groupby(['Entity', 'Code']).mean()
dirrahea_deaths_least_developed_avg

#The average of the five least developed countries dataframe.
```

```python
dirrahea_deaths_most_developed.to_csv("../sources/dirrahea_deaths_most_developed.csv", index=False, header=True)

dirrahea_deaths_least_developed.to_csv("../sources/dirrahea_deaths_least_developed.csv", index=False, header=True)
```

```python
dirrahea_deaths_most_developed_avg.to_csv("../sources/dirrahea_deaths_most_developed_avg.csv", index=False, header=True)

dirrahea_deaths_least_developed_avg.to_csv("../sources/dirrahea_deaths_least_developed_avg.csv", index=False, header=True)
```
