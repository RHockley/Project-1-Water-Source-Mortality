```python
bar_plot_unsafe = dirrahea_deaths_least_developed_tidy_avg[["Country", "Deaths - Cause: Diarrheal diseases - Risk: Unsafe water source - Sex: Both - Age: Under 5"]]

bar_plot_unsafe.plot(kind = "bar", color = "orange")

plt.xticks(np.arange(5), ['Burundi', 'Madagascar', 'Malawi', 'Mozambique', 'Somalia'], rotation = '45')
plt.title("The Five Least Developed Countries Diarrheal Deaths Caused by Unsafe Water Sources for Ages Under 5")
plt.ylabel("Diarrheal Deaths")
plt.xlabel("Five Least Developed Countries")
plt.legend('',frameon=False);
```

```python
bar_plot_unsafe = dirrahea_deaths_least_developed_tidy_avg[["Country", "Deaths - Cause: Diarrheal diseases - Risk: No access to handwashing facility - Sex: Both - Age: Under 5"]]

bar_plot_unsafe.plot(kind = "bar", color = "#CCCCFF")

plt.xticks(np.arange(5), ['Burundi', 'Madagascar', 'Malawi', 'Mozambique', 'Somalia'], rotation = '45')
plt.title("The Five Least Developed Countries Diarrheal Deaths Caused by No Handwashing Facilities for Ages Under 5")
plt.ylabel("Diarrheal Deaths")
plt.xlabel("Five Least Developed Countries")
plt.legend('',frameon=False);
```

```python
most_developed_nhf_pop = dirrahea_deaths_most_developed_tidy_avg["Deaths - Cause: Diarrheal diseases - Risk: No access to handwashing facility - Sex: Both - Age: Under 5"]
least_developed_nhf_pop = dirrahea_deaths_least_developed_tidy_avg["Deaths - Cause: Diarrheal diseases - Risk: No access to handwashing facility - Sex: Both - Age: Under 5"]
st.ttest_ind(most_developed_nhf_pop, least_developed_nhf_pop, equal_var = False)
```
