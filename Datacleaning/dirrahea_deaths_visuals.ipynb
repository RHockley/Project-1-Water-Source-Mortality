```python
labels = ["Burundi", "Madagascar", "Malawi", "Mozambique", "Somalia"]

sizes = [5037, 10662, 2628, 3960, 7950]

colours = ["orange", "red", "yellow", "gold", "brown"]

explode = (0, 0.1, 0, 0, 0)
```
